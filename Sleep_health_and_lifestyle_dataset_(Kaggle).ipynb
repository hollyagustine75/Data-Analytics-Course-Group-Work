Data Preparation / Preprocessing

```python
# Cek data null
df.isna().sum()
```

```python
#Hapus data null (jika ada)
df.dropna(inplace=True)
```

```python
#Cek data duplikasi
df.duplicated().sum()
```

```python
#Hapus data duplikat (jika ada)
df.drop_duplicates()
```

```python
#pemisahan Blood pressure ke systolic BP dan Diastolic BP

df[['Systolic BP', 'Diastolic BP']] = df['Blood Pressure'].str.split('/', expand=True)

# Convert the new columns to numeric type
df[['Systolic BP', 'Diastolic BP']] = df[['Systolic BP', 'Diastolic BP']].apply(pd.to_numeric)

# Drop the original 'Blood Pressure' column
df = df.drop('Blood Pressure', axis=1)
```

Data Analysis & Visualizations

```python
# @title Gender

plt.figure(figsize=(6, 6))
df['Gender'].value_counts().plot(kind='pie', colors=['blue', 'pink'], autopct='%1.1f%%')
plt.ylabel('')
plt.title('Gender Distribution')
```

```python
# @title Age

df['Age'].plot(kind='hist', bins=20, title='Age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Sleep Duration

df['Sleep Duration'].plot(kind='hist', bins=20, title='Sleep Duration')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Quality of Sleep

df['Quality of Sleep'].plot(kind = 'hist', bins = 20, title = 'Quality of Sleep')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title BMI Category

plt.figure(figsize=(6, 6))
df['BMI Category'].value_counts().plot(kind='pie', colors=sns.palettes.mpl_palette('Dark2'), autopct='%1.1f%%')
plt.ylabel('')
plt.title('BMI Category')
```

```python
# @title Sleep Disorder

plt.figure(figsize=(6, 6))
df['Sleep Disorder'].value_counts().plot(kind='pie', colors=sns.palettes.mpl_palette('Dark2'), autopct='%1.1f%%')
plt.ylabel('')
plt.title('Sleep Disorder')
```

```python
# @title Systolic BP

df['Systolic BP'].plot(kind = 'hist', bins = 20, title = 'Systolic BP')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Diastolic BP

df['Diastolic BP'].plot(kind = 'hist', bins = 20, title = 'Diastolic BP')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title Gender vs Quality of Sleep

plt.figure(figsize=(6, 1.2 * len(df['Quality of Sleep'].unique())))
sns.barplot(data=df, x='Gender', y='Quality of Sleep', palette='Dark2')
plt.title('Gender vs Quality of Sleep')
sns.despine(top=True, right=True, left=True, bottom=True)
plt.show()
```

```python
# @title Gender vs Sleep Duration

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 1.2 * len(df['Sleep Duration'].unique())))
sns.barplot(data=df, x='Gender', y='Sleep Duration', palette='Dark2')
plt.title('Gender vs Sleep Duration')
sns.despine(top=True, right=True, left=True, bottom=True)
plt.show()
```
